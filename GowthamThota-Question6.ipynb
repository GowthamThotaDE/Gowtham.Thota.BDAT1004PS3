# Question 6

Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.


```python
import pandas as pd
import matplotlib.pyplot as plt

us_marriage_divorce = pd.read_csv('us-marriages-divorces-1867-2014.csv')
years = us_marriage_divorce['Year'].values
divorce_cap = us_marriage_divorce['Divorces_per_1000'].values
marriage_cap = us_marriage_divorce['Marriages_per_1000'].values

plt.plot(years,marriage_cap, years,divorce_cap,'m')
plt.title('Line plot on the number of weddings and divorces per capita in the US between 1867 and 2014')
plt.xlabel('years')
plt.ylabel('marriage_cap and divorce_cap')
plt.legend(["marriage_cap", "divorce_cap"], loc ="lower right")
plt.show()
```

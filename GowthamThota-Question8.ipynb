Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort
the actors by their kill count and label each bar with the corresponding actor's name.

```python
import pandas as pd
import matplotlib.pyplot as plt

hollywood_actor_kill_counts = pd.read_csv('actor_kill_counts.csv')
hollywood_actor_kill_counts_descending = hollywood_actor_kill_counts.sort_values('Count',ascending=False)

actors = hollywood_actor_kill_counts_descending['Actor'].values
kills_count = hollywood_actor_kill_counts_descending['Count'].values

plt.barh(actors,kills_count)
plt.title('Horizontal Bar Plot of Hollywood Actors kill count')
plt.xlabel('kills_count')
plt.ylabel('actors')
plt.show()
```

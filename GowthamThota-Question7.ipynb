# Question-7

Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

us_marriage_divorce = pd.read_csv('us-marriages-divorces-1867-2014.csv')
us_marriage_divorce = us_marriage_divorce[us_marriage_divorce['Year'].apply(lambda x: x in [1900, 1950, 2000])]

years = us_marriage_divorce['Year'].values
divorce_cap = us_marriage_divorce['Divorces_per_1000'].values
marriage_cap = us_marriage_divorce['Marriages_per_1000'].values

p = np.arange(len(years))
plt.bar(p, marriage_cap, color = 'c', width = 0.3)
plt.bar(p + 0.3, divorce_cap , color = 'y', width = 0.3)
plt.legend(['Marriages per capita', 'Divorces per capita'])

plt.xticks([i + 0.3 for i in range(3)], [1900, 1950, 2000])
plt.title("Bar plot on the no.of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000")
plt.xlabel('years')
plt.ylabel('count of marraiges and divorces ')

plt.show()
```

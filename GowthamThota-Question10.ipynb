Create a scatter plot showing the relationship between the total revenue earned by
arcades and the number of Computer Science PhDs awarded in the U.S. between
2000 and 2009.

```python
import matplotlib.pyplot as plt
import pandas as pd

arcade_revenue_cs_phd = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')
arcade_revenue_b = arcade_revenue_cs_phd['Total Arcade Revenue (billions)'].values
phd_awarded = arcade_revenue_cs_phd['Computer Science Doctorates Awarded (US)'].values
years = arcade_revenue_cs_phd['Year'].values
arcade_revenue_cs_phd

plt.scatter(arcade_revenue_b, phd_awarded, alpha=0.5, c=years, edgecolors='none', s=50,cmap='Reds',linewidth = 6)
plt.title('Scatter plot between total arcade income and the no.of Computer Science PhDs in the U.S between 2000 and 2009')
cbar = plt.colorbar()

cbar.set_label('Years')
plt.ylabel('arcade revenue_b')
plt.xlabel('phd awarded')
plt.show()
```

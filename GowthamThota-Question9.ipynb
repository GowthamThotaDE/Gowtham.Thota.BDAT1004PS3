# Question-9

Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.

```python
import pandas as pd
import matplotlib.pyplot as plt

roman_emperor_reign = pd.read_csv('roman-emperor-reigns.csv')
assassinated_roman_emperors = roman_emperor_reign[
    roman_emperor_reign['Cause_of_Death'].apply(lambda x: 'assassinated' in x.lower())]


emperors_assassinated = len(assassinated_roman_emperors)
other_deaths = len(roman_emperor_reign) - emperors_assassinated


emperors = assassinated_roman_emperors["Emperor"]
cause_of_death = assassinated_roman_emperors["Cause_of_Death"]
plt.pie(range(len(cause_of_death)), labels=emperors, autopct='%1.1f%%', startangle=280, radius=0.04 * 60, rotatelabels = 60)
fig = plt.figure(figsize=[10, 10])
```
